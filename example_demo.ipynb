```python
from color_by_number import ColorByNumber
from pathlib import Path

path = Path("/Users/germanshepherd/PycharmProjects/color_by_number")
file_name = "test_1.JPG"

number_of_colors = 10
number_of_blocks = 100
```

```python
cbn = ColorByNumber(
    fig_path=path / "original_figures" / file_name,
    number_of_colors=number_of_colors,
    number_of_blocks=number_of_blocks,
    save_path=path / "outputs"
)

# Puzzle version: white blocks + numbers
puzzle = cbn.generate_mosaic(block_size=30)

# Separate legend: number -> color
legend = cbn.generate_reference_color_bar(columns=2)

# Revealed version: colored blocks
finished = cbn.complete_mosaic(block_size=30, show_numbers=False)

# Save all generated outputs
cbn.save_mosaic(
            mosaic_name=f"{file_name}_mosaic_puzzle.png",
            reference_name=f"{file_name}_reference_color_bar.png",
            completed_name=f"{file_name}_mosaic_completed.png",
)
```
